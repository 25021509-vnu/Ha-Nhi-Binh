```python
n = int(input(""))
if n > 2 :
  count = 0
  for i in range(1, n+1):
     if n % i == 0:
        count += 1
  if count == 2:
     print(f"{n} la so nguyen to")
  else:
     print(f"{n} khong la so nguyen to") 
```
